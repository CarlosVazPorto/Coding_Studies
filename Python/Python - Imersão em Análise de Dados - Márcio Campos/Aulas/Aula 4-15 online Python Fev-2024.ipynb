```python
ano_nascto  = "1968"       # mnemônico (dicionário) nascto = nascimento  
divida      = -100
idade       = 55
salario     = 1000
nota        = 9
notas       = [10, 8.6, 7, 3, 4.5, 9.3] # Estrutura de dados List, somente com inteiros e float

print("Divida      :", divida, abs(divida), type(divida))            # valor absoluta, sem alterar o tipo da variável
print("Idade       :", idade, idade * 2, type(idade))                # operação sem alterar o tipo da variável
print("Salario     :", salario, float(salario),type(salario))        # conversão de tipo, sem alterar o tipo da variável
print("Nota        :", nota, pow(nota,2),type(nota))                 # exponenciação sem alterar o tipo da variável
print("Nascto      :", ano_nascto, int(ano_nascto),type(ano_nascto)) # conversão de tipo, sem alterar o tipo da variável
print("Resto       :", salario, divmod(salario, 50))                 # quociente e resto, sem alterar o tipo da variável
print("Arredondar  :", salario, salario/3.2, round(salario/3.2))     # arredondamento, sem alterar o tipo da variável
print("Máximo      :", notas, type(notas), max(notas))               # máximo, sem alterar o tipo de dado da estrutura
print("Máximo      :", notas, type(notas), min(notas))               # minimo, sem alterar o tipo de dado da estrutura
print("Elemento    :", notas, int(round(notas[1])))                  # abs pontual, sem alterar o tipo de dado da estrutura
print("Somar       :", notas, type(sum(notas)), type(notas))         # somatorio, sem alterar o tipo de dado da estrutura
```

```python
# Um assunto chamado data
# -----------------------

# importa a classe datetime que é usada para representar datas e horas em Python
# ----------------------------------------------------------------------------------------------------------------------- #
from datetime import datetime

# importa a classe timedelta que é usada para representar diferenças entre datas e horas.
# ----------------------------------------------------------------------------------------------------------------------- #
from datetime import timedelta

# importa relativedelta, é uma classe que oferece funcionalidades avançadas para manipulação de datas e tempos em Python.
# Ela é especialmente útil para adicionar ou subtrair meses, anos, semanas, etc., de objetos datetime, 
# levando em consideração as diferenças nos comprimentos dos meses e anos.
# ----------------------------------------------------------------------------------------------------------------------- #
from dateutil.relativedelta import relativedelta

# strptime ("String Parse Time"):

# Este método é usado para converter strings em objetos datetime.
# Ele aceita dois argumentos: a string contendo a data e a hora a serem convertidas e um formato que especifica como
# a string está estruturada.
# O método strptime analisa a string de entrada de acordo com o formato especificado e retorna um objeto datetime
# correspondente.

aniversario       = "01/01/1968"

print("Aniversário       :", aniversario, datetime.strptime(aniversario, "%d/%m/%Y"), type(aniversario))

aniversario_primo = datetime.strptime(aniversario, "%d/%m/%Y")

print("Aniversário Primo :", aniversario_primo, type(aniversario_primo))
```

```python
# strftime ("String Format Time"):

# Este método é usado para converter objetos datetime em strings formatadas, permitindo que você especifique o formato
# exato da string de saída.
# Ele aceita um argumento de formato, que é uma string que define como você deseja que a data e a hora sejam formatadas
# na string resultante.
# O método strftime retorna uma string formatada de acordo com o formato especificado.

print("Aniversário Primo :", aniversario_primo, datetime.strftime(aniversario_primo,"%d/%m/%Y %H:%M:%S"), type(aniversario_primo))

aniversario_tia = datetime.strftime(aniversario_primo,"%d/%m/%Y %H:%M:%S")

print("Aniversário Tia :", aniversario_tia+" (Dita)", type(aniversario_tia))
```

```python
# Vamos acrescentar dias, meses e anos a data ?
# ---------------------------------------------

# Para acrescentar dias, meses ou anos a um objeto datetime em Python, você pode usar a classe timedelta
# do módulo datetime.

# Acrescentando dia, meses e anos a uma variável Data
# ---------------------------------------------------

print("Criando a variável datatime")

data_referencia = datetime(2024, 2, 18) 

print("Referencia : ", data_referencia, type(data_referencia))

# timedelta(days=7) acrescenta 7 dias à data de referência.
# timedelta(weeks=4) acrescenta aproximadamente 1 mês (considerando 4 semanas).
# timedelta(days=365) acrescenta aproximadamente 1 ano (considerando 365 dias).

print("    7 dias : ", data_referencia + timedelta(days=7), type(data_referencia))
print(" 4 semanas : ", data_referencia + timedelta(weeks=4), type(data_referencia))
print("     1 ano : ", data_referencia + timedelta(days=365), type(data_referencia))

# Para adicionar um mês a um objeto datetime em Python, precisa considerar algumas situações especiais devido
# às diferenças no número de dias em cada mês. A maneira mais comum de lidar com isso é usar a biblioteca
# dateutil.relativedelta, que oferece uma maneira simples e eficaz de manipular datas e tempos em Python.
# ----------------------------------------------------------------------------------------------------------------------- #

data_futura = data_referencia + relativedelta(months=1)

print("     1 mês : ", data_futura, type(data_referencia))
```

```python
# ----------------------------------------------------------------------------------------------------------------------- # 
# Variáveis Locais:
# Variáveis definidas dentro de uma função têm escopo local.
# Elas só podem ser acessadas dentro da função onde foram definidas.
# As variáveis locais são criadas quando a função é chamada e são destruídas quando a função retorna.
# Tentar acessar uma variável local fora do escopo da função onde foi definida resultará em um erro.
# ---------------------------------------------------------------------------------------------------------------------- #

# ---------------------------------------------------------------------------------------------------------------------- # 
# Variáveis Globais:
# Variáveis definidas fora de qualquer função têm escopo global.
# Elas podem ser acessadas de qualquer lugar do código, incluindo dentro de funções.
# Se uma variável global é modificada dentro de uma função, ela precisa ser declarada como global dentro da função.
# Variáveis globais são criadas quando são atribuídas um valor pela primeira vez e existem até o programa terminar
# ou até serem excluídas explicitamente.
# ---------------------------------------------------------------------------------------------------------------------- #

"""
    Funções do Python para descobrir se a variável é global ou local
        globals() / locals()
    
    Melhores práticas : desenvolver com documentação, lógica e algoritimo estruturado
"""

# (escopo global)
# ---------------

print("Global")

x = 10

def AulaVariavel():
    global x
    x += 10
    print(x)

AulaVariavel()

print(x)

# (escopo local)
# ---------------
print("Local")
y = 10

def AulaVariavel():    
    y = 5 
    print(y)

AulaVariavel()

print(y)
```

```python
# ----------------------------------------------------------------------------------------------------------------------- #
# Os métodos rsplit() e split() são usados para dividir uma string em substrings com base em um separador e 
# retornar uma lista das substrings resultantes. A principal diferença entre eles está na direção em que a string é 
# dividida.

# O método split() divide uma string em substrings usando um separador especificado
# e retorna uma lista das substrings resultantes.

# O método rsplit() é semelhante ao split(), mas divide a string da direita para a esquerda.
# ----------------------------------------------------------------------------------------------------------------------- #

frase = "Hoje é dia de aprender fundamentos e conceitos de Python"
print(frase.split())
print(frase.rsplit())

texto = frase.split(maxsplit=2)
print(texto)

linguagens = "C/C++ - Cobol - PHP - Python - Javascript - Node.JS - Rust Julia"
texto      = linguagens.split("-") 
print(texto)

print(frase)
```
