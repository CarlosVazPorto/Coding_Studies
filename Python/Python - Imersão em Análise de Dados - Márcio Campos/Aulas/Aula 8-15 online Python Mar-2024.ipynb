datetime[ns] é um tipo de dados utilizado pelo pandas para representar datas e horas com precisão de nanossegundos. A sigla ns refere-se a nanossegundos, onde um nanossegundo é um bilionésimo de segundo (1 segundo = 1.000.000.000 nanossegundos). Esse tipo de dados é especialmente útil para manipulação de séries temporais em análises financeiras, científicas, ou qualquer outro domínio onde a precisão temporal em alta resolução é necessária.

Características de datetime[ns]
Alta Precisão: O uso de nanossegundos permite uma representação muito precisa de timestamps, útil para dados que requerem marcações temporais detalhadas, como transações financeiras ou eventos em sistemas de alta frequência.

Compatibilidade com NumPy: Pandas utiliza o NumPy por baixo dos panos, que suporta datetime64[ns]. Isso permite integração e operações eficientes com arrays de NumPy e facilita a manipulação de dados temporais em larga escala.

Operações Temporais: Com o tipo datetime[ns], pandas permite realizar operações como diferenças de tempo, resampling (reagrupamento de dados temporais), shifting (deslocamento de dados no tempo), entre outras, com facilidade e precisão.

Um lambda no Python é basicamente uma forma curta de definir uma função anônima, ou seja, uma função sem nome. Ela pode ter qualquer número de argumentos, mas só pode ter uma expressão. Lambdas são úteis quando você precisa de uma função simples para uma operação de curta duração e não quer formalmente defini-la com o def.

```python
quadrado_lambda = lambda num: num ** 2
print(quadrado_lambda(11))
```

A função map aplica uma função especificada a cada item de um iterável (como uma lista) e retorna um novo iterável com os resultados. map é comumente usado quando você quer transformar os dados de uma forma para outra.

```python
numeros   = [1, 2, 3, 4, 5]
quadrados = list(map(lambda num: num ** 3, numeros))
print(quadrados)
```

A função filter aplica uma função especificada, que deve retornar True ou False, a cada item de um iterável. filter retorna um novo iterável contendo apenas os itens para os quais a função aplicada retorna True. É útil quando você precisa filtrar um conjunto de dados de acordo com algum critério.

```python
numeros = [1, 2, 3, 4, 5]
maiores_que_dois = list(filter(lambda num: num >= 2, numeros))
print(maiores_que_dois)
```

```python
def ajusta_virgula(coluna):
    
    if pd.isna(coluna):
        return 0
    
    elif isinstance(coluna, str):
        try:
            return float(coluna.replace(',', '.'))
        except ValueError:
            return 0
    
    else:
        return coluna

ajusta_colunas = ['Fï¿½rias', 'CestaBï¿½sica', 'AjudaCusto']

for coluna in ajusta_colunas:
    df_folha[coluna] = df_folha[coluna].apply(lambda x: ajusta_virgula(x))

df_folha.rename(columns={'Fï¿½rias': 'Ferias', 'CestaBï¿½sica': 'CestaBasica'}, inplace=True)
```

O termo axis (eixo) em bibliotecas de manipulação de dados como pandas e numpy refere-se à direção ao longo da qual uma operação é realizada em uma estrutura de dados. Em estruturas de dados multidimensionais como DataFrames (pandas) e arrays (numpy), axis ajuda a especificar se você deseja realizar uma operação ao longo das linhas ou colunas (ou em outras dimensões, no caso de arrays com mais de duas dimensões).

```python
tem_nan_linhas = df_folha.isna().any(axis=1).any()

print(tem_nan_linhas)
```

```python
tem_nan_colunas = df_folha.isna().any(axis=0)

print(tem_nan_colunas)
```

```python
caracteres_validos = r'[^a-záàéèíìóòúùA-ZÁÀÉÈÍÌÓÒÚÙçÇ@,.-0-9]'

mascara            = df_folha.applymap(lambda x: bool(re.search(caracteres_validos, str(x))))

linhas_com_caractere_especial = mascara.any(axis=1).sum()
colunas_com_caractere_especial = mascara.any().sum()

print(f'Linhas com caractere especial: {linhas_com_caractere_especial}')
print(f'Colunas com caractere especial: {colunas_com_caractere_especial}')
```

applymap(lambda x: bool(re.search(caracteres_validos, str(x)))): Aplica uma função a cada elemento do DataFrame que verifica se o elemento contém qualquer caractere fora do conjunto definido de caracteres válidos. Isso cria uma máscara booleana onde True indica a presença de um caractere especial.

mascara.any(axis=1).sum(): Calcula o número de linhas que contêm pelo menos um True na máscara booleana, ou seja, pelo menos um caractere especial.

mascara.any().sum(): Calcula o número de colunas que contêm pelo menos um True.
